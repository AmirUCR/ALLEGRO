```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

increments = list(dict.fromkeys((np.geomspace(1, 2434, 30).astype(int))))
```

```python
import re

plt.rcParams['figure.dpi'] = 600

a1_dirs = [f'track_a1_ns_{i}species' for i in increments]
a1_data = dict()

for d in a1_dirs:
    species_match = re.match(r'track_a1_ns_(\d+)species', d)
    n_species = int(species_match.group(1))

    if os.path.exists(f'{d}/{d}.csv'):
        size = min(pd.read_csv(f'{d}/trial_sizes.csv')['library_size'].values.tolist())
        a1_data[n_species] = size
    else:
        a1_data[n_species] = 0

# -----

a7_dirs = [f'track_a7_ns_{i}species' for i in increments]
a7_data = dict()

for d in a7_dirs:
    species_match = re.match(r'track_a7_ns_(\d+)species', d)
    n_species = int(species_match.group(1))

    if os.path.exists(f'{d}/{d}.csv'):
        size = min(pd.read_csv(f'{d}/trial_sizes.csv')['library_size'].values.tolist())
        a7_data[n_species] = size
    else:
        a7_data[n_species] = 0

# ----

e1_dirs = [f'track_e1_ns_{i}species' for i in increments]
e1_data = dict()

for d in e1_dirs:
    species_match = re.match(r'track_e1_ns_(\d+)species', d)
    n_species = int(species_match.group(1))

    if os.path.exists(f'{d}/{d}.csv'):
        size = min(pd.read_csv(f'{d}/trial_sizes.csv')['library_size'].values.tolist())
        e1_data[n_species] = size
    else:
        e1_data[n_species] = 0
```

```python
plt.figure(figsize=(12, 6))
plt.plot(a1_data.keys(), a1_data.values(), color='#ff2647')
for key, value in list(a1_data.items())[0:1]:
    plt.text(key, value + 10, str(value), ha='center', va='bottom', fontsize=8)

for key, value in list(a1_data.items())[14:15]:
    plt.text(key, value + 7, str(value), ha='center', va='bottom', fontsize=8)

for key, value in list(a1_data.items())[15:16]:
    plt.text(key, value + 14, str(value), ha='center', va='bottom', fontsize=8)

for key, value in list(a1_data.items())[16:21]:
    plt.text(key, value + 7, str(value), ha='center', va='bottom', fontsize=8)

for key, value in list(a1_data.items())[21:]:
    plt.text(key, value + 5, str(value), ha='center', va='bottom', fontsize=8)
    
plt.xlabel('Number of Species')
plt.ylabel('Smallest Library Size in 2,000 Trials')
plt.title('Smallest Library Size vs. Number of Species for Track A - M1')
plt.xticks([*a1_data.keys()], rotation=45)

ax = plt.gca()
xticks = ax.xaxis.get_major_ticks() 
for i in range(1, 14):
    xticks[i].label1.set_visible(False)

plt.grid(alpha=0.1, linestyle='--')
plt.tight_layout()
plt.show()
```
