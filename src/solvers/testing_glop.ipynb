```python
from ortools.linear_solver import pywraplp
solver = pywraplp.Solver.CreateSolver('GLOP')

S = {
    's1': ['x1', 'x2', 'x3'],
    's2': ['x2', 'x4'],
}

X = {
    'x1': [80, {1}],
    'x2': [60, {1, 2}],
    'x3': [70, {1}],
    'x4': [90, {2}]
}

beta = 1

vars = dict((x, solver.NumVar(lb=0.0, ub=1.0, name=x)) for x in X.keys())

for species in S.keys():
    solver.Add(solver.Sum([vars[seq] for seq in S[species]]) >= 1)

objective_1_terms = list()
objective_2_terms = list()

for seq, var in vars.items():
    objective_1_terms.append(1 * var)
    objective_2_terms.append(X[seq][0] * var)

objective_1 = solver.Sum(objective_1_terms)
objective_2 = solver.Sum(objective_2_terms)

solver.Add(objective_1 <= beta)

solver.Maximize(objective_2)


solver.Solve()
solutions = list()
for seq, var in vars.items():
    if var.solution_value() > 0.0:
        print('Guide: {seq} \t with probability: {p:.2f} \t with score: {score} \t covering species: {species}'.format(seq=seq, p=var.solution_value(), score=X[seq][0], species=X[seq][1]).expandtabs(3))

        solutions.append(var)

print(solver.ExportModelAsLpFormat(False))
```
