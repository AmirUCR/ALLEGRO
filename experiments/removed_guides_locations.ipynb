```python
import pandas as pd

df = pd.read_csv('../removed_guides.csv')

df.head()
```

```python
scer = df[df['species'] == 'saccharomyces_cerevisiae']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(15, 6))

sns.violinplot(
    x='seq_id',
    y='proportion_of_total_location',
    data=df[df['species'] == 'saccharomyces_cerevisiae']
    )

plt.ylim(0, 1)
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(15, 6))

sns.violinplot(
    x='seq_id',
    y='proportion_of_total_location',
    data=df[(df['species'] == 'saccharomyces_cerevisiae') & ((df['contains_fourmers'] == 1) | (df['contains_fivemers'] == 1))]
    )

plt.ylim(0, 1)
plt.xticks(rotation=90)
# Show the plot
plt.show()
```
